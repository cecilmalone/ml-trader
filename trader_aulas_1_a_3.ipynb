```python
import matplotlib.pyplot as plt
# importa a biblioteca para plotagens

fig = plt.figure(figsize=(10,10))
# cria a figura com dimenses 10 x 10

ax = fig.gca()
# cria os vrtices da figura

a = [1, 2, 3, 4, 5]
# Coluna horizontal da plotagem, representando os dias

b = [11, 23, 42, 17, 25]
# Coluna vertical da plotagem, representando o preo

ax.plot(a, b, label = "teste")
# plota os valores a & b com o label 'teste'

plt.legend()
# habilita as legendas

plt.show()
# Exibe a plotagem
```
